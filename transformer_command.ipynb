```python
qa = pipeline('question-answering')
```

```python
c = '''Bity brought some butter, but the butter was bitter butter.
so,Bitty bought some better butter to make the bitter butter better'''
```

```python
q = "What does Bitty bought?"
```

```python
result = qa({'question':q,'context':c})
print(result)
```
