```python
import pydub
from pydub import AudioSegment
mp3_file = "/content/Doraemon.mp3"
wav_file = "/content/Doraemon.wav"
Audio = AudioSegment.from_mp3(mp3_file)
Audio.export(wav_file,format="wav")
```

```python
import speech_recognition as sr
reco = sr.Recognizer()
with sr.AudioFile("/content/Doraemon.wav") as source:
  audio_data = reco.record(source)
  print("Transcript:",reco.recognize_google(audio_data))
```

```python
from PIL import Image
import pytesseract
import matplotlib.pyplot as plt
ip = "/content/quote 4.jpg"
plt.imshow(Image.open(ip))
plt.axis('off')
```

```python
img = Image.open(ip)
text = pytesseract.image_to_string(img)
print("Extracted text:",text)
```
